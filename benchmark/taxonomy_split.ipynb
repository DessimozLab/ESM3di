# Split dataset based on taxonomy

## Split MetaVR dataset

```python
%%bash
cat clades.txt | while read tax tlc; do
    awk '\
        NR==FNR {f[">"$1]=1; next} \
        $1 in f {print; getline; print}' \
    <(grep $tax data/metavr/metavr_taxonomy.tsv | cut -f1 | awk '{print tolower($0)}') \
    data/metavr/metavr_aa.fa \
    > data/metavr/metavr_${tlc}_aa.fa
done
```
