# Extract taxonomy from the MetaVR dataset

## Description

MetaVR AlphaFold3 predictions were performed on the representatives of the protein clusters derived from the entire MetaVR dataset.

Here we extract the taxonomy mapping from the MetaVR protein cluster representatives to the ICTV taxonomy.

The final table will have the following structure:
```
Cluster_ID \t UViG_ID \t ICTV_taxonomy
```
For example:
```
IMGVR_PC_000000001 \t IMGVR_UViG_3300009285_008502 \t r__Duplodnaviria;k__Heunggongvirae...
```

## Download and decompress relevant files

````python
%%bash
wget -q -O data/metavr/metavr_uvig_metadata.tsv.gz "https://www.meta-virome.org/DownloadUvigMetadata"
tar -xzf data/metavr/IMGVR5_PC_3Dmodels.tar.gz -C data/metavr
wget -q -O data/metavr/IMGVR5_PC_MSAs.tar.zst "https://www.meta-virome.org/Data/Downloads/IMGVR5_PC_MSAs.tar.zst"
tar --zstd -xf data/metavr/IMGVR5_PC_MSAs.tar.zst -C data/metavr
````

## Parse files to extract mapping

````python
%%bash
ls -U data/metavr/MSAs | while read F; do
    CID=${F%%.*}
    # first one is the representative
    VID=$(head -n1 data/metavr/MSAs/${F} | awk -F'|' '{print substr($1, 2)}')
    echo -e "${CID}\t${VID}"
done > data/metavr/metavr_cluster_uvig.tsv

awk -F'\t' 'NR==FNR { f[$1] = $19; next } { print $0 "\t" f[$2] }' \
    data/metavr/metavr_uvig_metadata.tsv data/metavr/metavr_cluster_uvig.tsv \
    > data/metavr/metavr_taxonomy.tsv
````
