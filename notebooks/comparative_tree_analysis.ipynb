# Comparative Tree Analysis: Structure vs Sequence-Based 3Di

This notebook compares phylogenetic trees built using:
1. **Ground truth 3Di** - Extracted from protein structures using FoldSeek
2. **Predicted 3Di** - Inferred from amino acid sequences using ESM3Di model
3. **Traditional amino acid sequences**

## Tree Building Methods
- **Maximum Likelihood (RAxML)** - More accurate, slower
- **Distance-based (QuickTree)** - Faster, useful for large datasets

## Requirements
- Input: Directory of PDB/mmCIF structure files
- Tools: FoldSeek, MAFFT, RAxML-ng, QuickTree
- ESM3Di checkpoint for 3Di prediction

```python
import tempfile
import shutil
import os
import tqdm
def create_foldseek_db( struct_glob , dbpath , dbname="flavi_db"):
	"""Create a Foldseek database from a glob of structure files."""
	struct_paths = {}
	with tempfile.TemporaryDirectory() as tmpdir:
		# Create a temporary directory to hold the structure files
		for struct_folder in tqdm.tqdm(glob.glob(struct_glob)):
			pdbs = glob.glob(struct_folder + "/*.pdb")
			cifs = glob.glob(struct_folder + "/*.cif")
			if len(pdbs) + len(cifs) > 20:
				continue  # Skip folders with too many structures to avoid memory issues
			for struct_file in pdbs + cifs:
				struct_paths[os.path.basename(struct_file)] = struct_file
				shutil.copy(struct_file, tmpdir)
		# Create the Foldseek database
		run_foldseek_createdb(tmpdir, dbpath, dbname)
		return dbpath , struct_paths
```

```python
#use easy cluster on the database
def run_foldseek_easy_cluster(db, output_dir, foldseek_bin):
    """Run Foldseek easy-cluster on a database."""
    os.makedirs(output_dir, exist_ok=True)
    os.makedirs(os.path.join(output_dir, "temp"), exist_ok=True)

    if not os.path.exists(output_dir):
        cmd = [
            foldseek_bin,
            "cluster",
            db,
            output_dir,
            os.path.join(output_dir, "temp"),
            "-c", "0.01",
                ]
        subprocess.run(cmd, check=True)
    # createtsv DB DB DB_clu DB_clu.tsv
    
    cmd = [
        foldseek_bin,
        "createseqfiledb",
        db,
        output_dir,
        output_dir+"_clu_seq",
    ]
    subprocess.run(cmd, check=True)

    cmd = [
        foldseek_bin,
        "result2flat",
        db,
        db,
        output_dir+"_clu_seq",
        output_dir+"_clu_seq.fasta",
    ]
    subprocess.run(cmd, check=True)


    cmd = [
        foldseek_bin,
        "createtsv",
        db,
        db,
        output_dir,
        output_dir+"_clu.tsv",
    ]
    subprocess.run(cmd, check=True)

    print(f"Foldseek easy-cluster completed. Clusters saved to: {output_dir}")

run_foldseek_easy_cluster(os.path.join(OUTPUT_DIR, "flaviviridae_foldseek_db" , "flavi_db"), os.path.join(OUTPUT_DIR, "flaviviridae_clusters"), 'foldseek')
```

```python
#read the clustering output and print cluster sizes
clustering = os.path.join(OUTPUT_DIR, "flaviviridae_clusters_clu.tsv")
print( clustering )
#open with pandas and print cluster sizes
import pandas as pd
clu_df = pd.read_csv(clustering, sep="\t", header=None, names=["cluster_id", "seq_id", "target_cluster_id", "target_seq_id", "evalue", "bitscore"])
cluster_sizes = clu_df["cluster_id"].value_counts()
#select those larger than 10 
cluster_sizes = cluster_sizes[cluster_sizes > 10]
print("Cluster sizes:")
print(cluster_sizes)
#hist of cluster sizes
import matplotlib.pyplot as plt
plt.hist(cluster_sizes, bins=10, align='left')
plt.xlabel("Cluster Size")
plt.ylabel("Frequency")
plt.title("Distribution of Cluster Sizes from Foldseek Easy-Cluster")
plt.show()
```

```python
#create a folder for each cluster and copy the corresponding structures into it
for cluster_id in tqdm.tqdm(cluster_sizes.index):
    cluster_folder = os.path.join(OUTPUT_DIR, f"cluster_{cluster_id}")
    os.makedirs(cluster_folder, exist_ok=True)
    cluster_seqs = clu_df[clu_df["cluster_id"] == cluster_id]["seq_id"].tolist()
    for seq_id in cluster_seqs:
        #find the corresponding structure file in the original structure folder and copy it to the cluster folder
        struct_file = glob.glob(os.path.join(STRUCTURE_DIR, f"{seq_id}.*"))  # look for both .pdb and .cif
        if struct_file:
            shutil.copy(struct_file[0], cluster_folder)
```
